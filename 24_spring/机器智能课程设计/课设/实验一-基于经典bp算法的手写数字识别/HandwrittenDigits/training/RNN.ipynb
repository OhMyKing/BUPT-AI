```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam

# 加载数据
data = pd.read_csv('data/penbased-5an-nn.csv', header=None)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# 数据预处理
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

encoder = OneHotEncoder(sparse=False)
y_encoded = encoder.fit_transform(y.reshape(-1, 1))

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# RNN模型要求输入是3D的，我们需要增加一个时间维度
# 这里我们假设每个特征是序列的一步
X_train_rnn = np.expand_dims(X_train, axis=1)
X_test_rnn = np.expand_dims(X_test, axis=1)

# 定义RNN模型
model = Sequential([
    SimpleRNN(32, input_shape=(1, 16)), # 16维特征，假设每个特征是序列的一步
    Dense(10, activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.009), loss='categorical_crossentropy', metrics=['accuracy'])

# 训练模型
model.fit(X_train_rnn, y_train, epochs=60, batch_size=32, validation_split=0.2)

# 测试模型
test_loss, test_acc = model.evaluate(X_test_rnn, y_test)
print(f"Test accuracy: {test_acc:.4f}")
```
