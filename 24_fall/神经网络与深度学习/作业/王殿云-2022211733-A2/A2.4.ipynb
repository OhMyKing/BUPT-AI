# A2.4 在分类问题中，最小化交叉熵损失等价于最大化似然的推导

在机器学习的分类问题中，最小化交叉熵损失等价于极大化似然。下边给出这一结论的证明。

设数据集$\{(x_i, y_i)\}_{i=1}^N$，其中：

- $x_i$：输入特征向量
- $y_i$：目标输出（类别标签），$y_i \in \{1, 2, \dots, N\}$，$N$为类别数量

设模型的预测函数为$f(x_i; \theta)$，其中$\theta$为模型参数。对于多分类问题，我们通常使用**Softmax**函数将模型的输出转换为各类别的概率分布。对于输入$x_i$，模型的输出为一个实数向量$\mathbf{z}_i = [z_{i1}, z_{i2}, \dots, z_{iN}]^\top$。令$p_{iN}$表示样本$x_i$被预测为第$N$类的概率。通过Softmax函数，将其转换为概率分布：
$$
p_{iN} = \frac{\exp(z_{iN})}{\sum_{j=1}^N \exp(z_{ij})}
$$


## 一、似然函数

对于单个样本$x_i$，其真实类别为$y_i$，则其条件概率为：

$$
p(y_i \mid x_i; \theta) = p_{iy_i} = \frac{\exp(z_{iy_i})}{\sum_{j=1}^N \exp(z_{ij})}
$$

由于各个样本是独立的，整个数据集的似然函数为：

$$
L(\theta) = \prod_{i=1}^N p(y_i \mid x_i; \theta) = \prod_{i=1}^N \frac{\exp(z_{iy_i})}{\sum_{j=1}^N \exp(z_{ij})}
$$
$$
- \ln L(\theta) = - \sum_{i=1}^N \ln p(y_i \mid x_i; \theta) = - \sum_{i=1}^N \ln \left( \frac{\exp(z_{iy_i})}{\sum_{j=1}^N \exp(z_{ij})} \right)
$$
$$
- \ln L(\theta) = - \sum_{i=1}^N \left( \ln \exp(z_{iy_i}) - \ln \sum_{j=1}^N \exp(z_{ij}) \right ) = - \sum_{i=1}^N \left( z_{iy_i} - \ln \sum_{j=1}^N \exp(z_{ij}) \right )
$$

## 二、交叉熵损失函数

对于分类问题，交叉熵损失函数定义为：

$$
\text{CrossEntropy}(\theta) = - \sum_{i=1}^N \sum_{N=1}^N y_{iN} \ln p_{iN}
$$

其中，$y_{iN}$是样本$x_i$的真实标签的 one-hot 编码，若$y_i = N$，则$y_{iN} = 1$，否则$y_{iN} = 0$。

将$p_{iN}$代入，可以得到：

$$
\text{CrossEntropy}(\theta) = - \sum_{i=1}^N \ln p_{iy_i} = - \sum_{i=1}^N \left( z_{iy_i} - \ln \sum_{j=1}^N \exp(z_{ij}) \right )
$$

可以看出，表达式形式与负对数似然函数形式完全一致，可以认为两者等价。

综上所述，**在分类问题中，最小化交叉熵损失等价于极大化似然。。**
