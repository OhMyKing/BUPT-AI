# A2.3 在回归问题中，最小化均方误差等价于最小化负对数似然的推导

在机器学习的回归问题中，假设输出中包含高斯噪音，则最小化均方误差等价于最小化负log似然(或最大化似然)。下面给出这一结论的证明。


设数据集$\{(x_i, y_i)\}_{i=1}^N$，其中：

- $x_i$：输入特征向量
- $y_i$：目标输出（标量）

设模型预测函数为$f(x_i; \theta)$，其中$\theta$为模型参数。

设输出$y_i$包含高斯噪声$\epsilon_i$：

$$
y_i = f(x_i; \theta) + \epsilon_i
$$

其中，$\epsilon_i$是高斯噪声：

$$
\epsilon_i \sim \mathcal{N}(0, \sigma^2)
$$

机器学习的回归问题中，目标为估计模型参数$\theta$，使得模型对数据的拟合最好。由于$\epsilon_i$服从高斯分布，那么对于给定的$x_i$和$\theta$，$y_i$的条件概率密度函数为：

$$
p(y_i \mid x_i; \theta) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{(y_i - f(x_i; \theta))^2}{2\sigma^2} \right)
$$

由于各个样本是独立的，整个数据集的似然函数为：
$$
L(\theta) = \prod_{i=1}^N p(y_i \mid x_i; \theta) = \left( \frac{1}{\sqrt{2\pi \sigma^2}} \right)^N \exp\left( -\frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i; \theta))^2 \right)
$$

$$
\ln L(\theta) = -N \ln \left( \frac{1}{\sqrt{2\pi \sigma^2}} \right) + \frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i; \theta))^2
$$

$$
\ln L(\theta) = \frac{N}{2} \ln(2\pi \sigma^2) + \frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i; \theta))^2
$$

第一个项$\frac{N}{2} \ln(2\pi \sigma^2)$与$\theta$无关，因此在优化$\theta$时可以忽略，进一步得到：

$$
J(\theta) = \frac{1}{2\sigma^2} \sum_{i=1}^N (y_i - f(x_i; \theta))^2
$$

由于$\frac{1}{2\sigma^2}$是一个常数倍数，不影响优化的解，因此最小化$J(\theta)$等价于最小化均方误差（去掉常数倍数）：

$$
\text{MSE}(\theta) = \frac{1}{N} \sum_{i=1}^N (y_i - f(x_i; \theta))^2
$$

因此$J(\theta)$和$\text{MSE}(\theta)$在$\theta$上的最优解是一致的。

综上所述，可以证明， **在回归问题中，假设输出中包含高斯噪音，则最小化均方误差等价于最小化负log似然(或最大化似然)**
